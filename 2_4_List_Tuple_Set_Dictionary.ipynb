Types of Objects in Python:
1. Immutable Objects
2. Mutable Objects

Immutable Objects:
1. int
2. float
3. String
4. bool
5. tuple


Mutable Objects:
1. List
2. Set
3. Dictionary

```python
#list should be included in the square brackets
my_list = [1, 2, 3, 4, 5]
print(my_list)
type(my_list)
```

```python
#lists can have multiple data types
my_list = [2,3,1.8, "English", True]
print(my_list)
type(my_list)
```

Lists are Mutable--> Changeable



```python
#add elements to a list
my_list = [2,3,1.8, "English", True]
my_list.append(6)
print(my_list)
```

```python
# print the elements of the list using their index
print(my_list[0])
print(my_list[4])
```

```python
#lists allow duplicate values
list_1 = [1,2,3,4,5,12,2,3]
print(list_1)
```

```python
print(len(list_1))
```

```python
#initiating an empty list
list_2 = []
print(list_2)
```

```python
list_2.append(5)
print(list_2)
```

```python
#delete an item in a list
list_2 = [2, 3, 1.8, 'English', True, 6]
print(list_2)

del list_2[2]
print(list_2)
```

```python
#join the two lists
list_3 = [1,2,3,4,5]
list_4 = [6,7,8,9,10]

list_5 = list_3 + list_4
print(list_5)
```

Tuple

```python
tuple_1 = (2, 3, 4, 5)
print(tuple_1)
type(tuple_1)
```

```python
#tuple allows multiple data types
tuple_2 = (1,2,3.5, "Machine Learning", False)
print(tuple_2)
```

```python
# converting a list to tuple
my_list = [3,4,5,6]
print(my_list)

my_tuple = tuple(my_list)
print(my_tuple)
```

```python
print(my_tuple[0])
print(my_tuple[1])
```

Tuples are immutable-->Unchangeable

```python
my_tuple.append(6)
```

```python
print(len(my_tuple))
```

Set

```python
#set should be enclosed in Curly brackets
my_set = {1,2,3,4,5}
print(my_set)
type(my_set)
```

```python
print(my_set[0])
```

```python
# convert list to a set
list_5 = [1,2,3,4]

x = set(list_5)
print(x)
type(x)
```

```python
# set doesn't allow duplicate values
set_3 = {1,2,3,4,5,1,2,3}
print(set_3)
```

Dictionary

Key value pair

```python
my_dict = {'name':'Dabvid' ,'age':20, 'Country':'India'}
print(my_dict)
type(my_dict)
```

```python
print(my_dict['name'])
print(my_dict['age'])
print(my_dict['Country'])
```

```python
#dictionary does not allow duplicate values
dict_1 = {'name':'Dabvid' ,'age':20, 'Country':'India','name':'Dabvid' ,'age':20, 'Country':'India'}
print(dict_1)
```
