#Basic Data types in Python:
1. Integer
2. Floating Point
3. Complex
4. Boolean
5. String

```python
#integers
a = 8
print(a)
```

```python
type(a)
```

```python
#floating point
b = 2.3
print(b)
type(b)
```

```python
#complex numbers
c = 1 + 3j
print(c)
type(c)
```

Conversion of one data type to another

```python
#int to float
x = 10
print(x)
type(x)
```

```python
y = float(x)
print(y)
type(y)
```

```python
#float to int
x = 5.88
print(x)
type(x)
```

```python
y = int(x)
print(y)
type(y)
```
