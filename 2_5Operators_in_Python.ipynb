#1.Arithmetic Operators


```python
num_1 = 20
num_2 = 10

#addition
sum = num_1 + num_2
print('sum = ',sum)

#substraction
diff = num_1 - num_2
print("Difference = ", diff)

#multiplication
pro = num_1 * num_2
print("product = ",pro)

#division
div = num_1 / num_2
print("division = ",div)

#exponent
exp = num_1 ** num_2
print("exponent = ",exp)

#modulus
mod = num_1 % num_2
print("reminder = ",mod)
```

#3. Comparison Operators

```python
a = 5
b = 10

print(a == b) #equal to
print(a != b) #not equal to
print(a > b)  #greater than
print(a < b)  #less than
print(a <= b)
print(a >= b)
```

#4. Logical Operators:

#6. Membership Operators

```python
a = 5
b = 10

c = [1,2,3,4,5]
print(a in c)
print(b in c)
```
